```python
#examine the target varaiable 'dignosis' which is the variable to predict 
#get a count of diagnosis observations by type
data1.diagnosis.value_counts()
```

```python
#recode the target variable 'diagosis' to certain the labels 'malignant' and 'benign'
data1['diagnosis'].replace(['M','B'],['Malignant','Benign'])
```

```python
#Build the Confusion Matrix on the test and train
from sklearn.metrics import confusion_matrix
clf_cm_test=confusion_matrix(y_test,y_pred)
clf_cm_test
```

```python
#Obtaining Accuracy
accuracy=clf.score(x_test,y_test)
```
